### Dictionary

```python
customers = {
            'Alice': 25,
            'Bob': 30,
            'Charlie': 35
           }
```

```python
print(customers)
```

```python
customer1 = {'name': 'Alice', 'age': 30, 'email': 'alice@example.com'}
customer1['state'] = 'California'
print(customer1)
```

```python
customer2 = {'name': 'Bob', 'age': 25, 'email': 'bob@example.com'}
customer2['state'] = 'Texas'
print(customer2)
```

```python
customer3 = {'name': 'Charlie', 'age': 35, 'email': 'charlie@example.com'}
customer3['state'] = 'New York'
print(customer3)
```

```python
customer_list = [customer1, customer2, customer3]
print(customer_list)
```
